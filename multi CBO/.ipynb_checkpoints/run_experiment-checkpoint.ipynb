\begin{align*}
aL_1+(1-a)L_2 &= a\left(x_1^2+\frac{1}{4}x_2^2\right)+(1-a)\left(\frac{1}{4}(x_1-2)^2+(x_2-2)^2\right)\\
&= \frac{1+3a}{4}x_1^2-(1-a)x_1+1-a+\frac{4-3a}{4}x_2^2-4(1-a)x_2+4(1-a)
\end{align*}
Therefore, 
\begin{equation*}
X^*_a = \left(\frac{2(1-a)}{1+3a}, \frac{8(1-a)}{4-3a}\right)
\end{equation*}

```python
np.random.seed(1)

config_path = 'config2.yaml'
simname_sub = '250219Exp'
result_path = './result'
def update_yaml_config(file_path, key, value):
    with open(file_path, 'r') as yaml_file:
        config = yaml.safe_load(yaml_file)

    config[key] = value

    with open(file_path, 'w') as yaml_file:
        yaml.safe_dump(config, yaml_file)

def get_experiment_num(simname_sub):
    # List all entries in the result path
    entries = os.listdir(result_path)

    # Filter the entries to only include directories
    directories = [entry for entry in entries if os.path.isdir(os.path.join(result_path, entry))]

    # Print the directories
    i = 0
    while True:
        if simname_sub + '_' + str(i) in directories:
            i += 1
        else:
            break
    return i

for beta in [16]:
    i = get_experiment_num(simname_sub)
    update_yaml_config(config_path, 'beta', beta)
    update_yaml_config(config_path,'simname',simname_sub + '_' + str(i))
    
    stream = open('config2.yaml', 'r')
    config = yaml.full_load(stream) 
    print(config)


    simnum = 0 

    rets = pd.read_csv(config["retdatapath"], index_col=['Date'])
    retmean_t = rets.mean()
    retcov_t = rets.cov().values

    L1 = lambda x: x[0]**2 + x[1]**2/4
    L2 = lambda x: (x[0]-2)**2/4 + (x[1]-2)**2 + 10
    a = np.random.uniform(size=config['nump'])
    #a = np.floor(2*np.random.uniform(size=config['nump']))
    print(a)
    minimizer = np.zeros((config['nump'], 2))
    minimum = np.zeros(config['nump'])
    for i in range(config['nump']):
        minimizer[i] = [2*(1-a[i])/(1+3*a[i]), 8*(1-a[i])/(4-3*a[i])]
        minimum[i] = a[i] * L1(minimizer[i]) + (1 - a[i]) * L2(minimizer[i])
    plt.plot(np.linspace(0,config['nump']-1, config['nump']), minimum)
    plt.savefig('minimum_diff10')
    
    a = a.reshape(-1,1)
    
    
    while simnum < config['sim']: 
        model = new2_CBO_model(L1, L2, config)  # 이 때, __init__이 실행됨.
        model.make_path()
        x = 10 * np.random.rand(config['nump'], config['D'])-5
        nump = config['nump']
        S = np.zeros(config['maxiter'])
        E = np.zeros(config['maxiter'])
        A = np.zeros(config['maxiter'])
        LL = np.zeros(config['maxiter'])
        it = 0  # number of interations
        while it < config['maxiter']:
            print(f"\r{int(1.0*(simnum*config['maxiter']+it)/(config['maxiter']*config['sim'])*100)+1} %", end="")
            sss = 0
            for i in range(nump):
                for j in range(nump):
                    sss += np.linalg.norm(x[i]-minimizer[j])**2
            S[it] = sss
            s = 0
            for i in range(nump):
                s+= np.linalg.norm(x[i]-minimizer[i])**2
            E[it] = s
            x_bar = model.weighted_avg(x,a,L1,L2)
            ss = 0
            for i in range(nump):
                ss += np.linalg.norm(minimizer[i]-x_bar[i])**2
            A[it] = ss     
            ssss = 0
            for i in range(nump):
                ssss += a[i][0]*L1(x[i]) + (1-a[i][0])*L2(x[i]) - (a[i][0]*L1(minimizer[i]) + (1-a[i][0])*L2(minimizer[i]))
            LL[it] = ssss
            if config['model'] == 1:
                x = model.step(x, a, L1,L2) 
            else:
                x = model.step_weight(x, L)

            it += 1
            
           

            model.trace_func(x, lambda x: x, "coord")

        model.save_func("coord", simnum=simnum)
#         plt.scatter(np.linspace(0,config['maxiter']-1, config['maxiter']),E, s = 5, label = 'E_n')
#         plt.scatter(np.linspace(0,config['maxiter']-1, config['maxiter']),A, s = 5, label = 'A_n')
        plt.plot(np.linspace(0,19, 20),E[:20], label = 'E_n')
        plt.plot(np.linspace(0,19, 20),A[:20], label = 'A_n')
        plt.plot(np.linspace(0,19, 20),S[:20], label = 'S_n')
#         plt.plot(np.linspace(0,19,20), LL[:20] ,label = 'LL_n')
        plt.title('Exp'+str(simnum))
        plt.xlabel('iter')
        plt.ylabel('')
        plt.yscale('log')
        plt.legend()
        
        plt.savefig(str(simnum)+"_diff10")
        plt.clf()
        #model.trace_func(x, lambda x: model.best_loss(x, L), "best_final")
        #model.trace_func(x, lambda x: avg_L(x, L), "avg_final")
        #model.save_func("best_final", simnum=simnum)
        #model.save_func("avg_final", simnum=simnum)
        simnum += 1
    model.save_config()

    print("")
    print("Done!")
```
